```python
# Create features list including variables from var_0 to var_9
features_list = df[['var_0', 'var_1', 'var_2', 'var_3', 'var_4', 'var_5', 'var_6', 'var_7', 'var_8', 'var_9']]
```

```python
# Find correlations in features
features_list.corr()
```

```python
# Sort most correlated variables in features
features_list = features_list.unstack()
```

```python
lst = []
for i in features_list:
    if i > 0.9 and i < 1:
        lst.append(i)
```

```python
series = pd.Series(lst)

series = series.drop_duplicates()
unique_list = series.tolist()
unique_list.sort()
len(unique_list)
```

```python
# Show top 10 most correlated variables in features
unique_list[:-10:-1]
```

```python
# Remove outliers from dataset
window_size = 10
outliers_mask = pd.Series(False, index=df.index)
for i in df.columns:
    mean = df[i].ewm(window_size).mean()
    std = df[i].ewm(window_size).std()
    std[0] = 0 #the first value turns into NaN because of no data
    mean_plus_std = mean + 2*std
    mean_minus_std = mean - 2*std
    is_outlier = (df[i] > mean_plus_std) | (df[i] < mean_minus_std)
    outliers = df[is_outlier]
    outliers_mask = outliers_mask | is_outlier
print(outliers)
```

```python
filtered_df = df[~outliers_mask]
filtered_df
```
